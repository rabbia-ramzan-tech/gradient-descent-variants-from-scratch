```python
#STOCHASTIC GRADIENT DESCENT

true_w = 2
true_b = 1
N = 100
np.random.seed(100)
x = np.random.rand(N, 1)
epsilon = 0.1 * np.random.randn(N, 1)
y = true_w * x + true_b + epsilon
idx = np.arange(N)
np.random.shuffle(idx)
idx_train = idx[:int(0.8 * N)]
idx_test = idx[int(0.8 * N):]
x_train, y_train = x[idx_train], y[idx_train]
x_val, y_val = x[idx_test], y[idx_test]
# -----------------------------------------

learning_rates = [0.001, 0.01, 0.05, 0.1, 0.5]

print("STARTING STOCHASTIC GRADIENT DESCENT")

for lr in learning_rates:
    print(f"\n=== Running SGD for Learning Rate = {lr} ===")

    trainLosses = []
    valLosses = []

    # Initialize weights
    w = np.random.randn(1)
    b = np.random.randn(1)

    for i in range(100): # Epochs

        # Shuffle training data for SGD
        perm = np.random.permutation(len(x_train))
        x_shuffled = x_train[perm]
        y_shuffled = y_train[perm]

        # Iterate through EACH sample
        for j in range(len(x_train)):
            xi = x_shuffled[j:j+1]
            yi = y_shuffled[j:j+1]

            # Forward pass (single sample)
            yhat_i = w * xi + b
            error_i = yhat_i - yi

            # Gradients (single sample)
            db = 2 * error_i
            dw = 2 * xi * error_i

            # Update immediately
            b = b - lr * db
            w = w - lr * dw

        # --- Calculate Full Epoch Loss for plotting ---
        # (We evaluate the "current" w and b on the full set to generate the curve)
        yhat_full = w * x_train + b
        loss = ((yhat_full - y_train)**2).mean()
        trainLosses.append(loss)

        # Validation
        yhatVal = w * x_val + b
        errorVal = yhatVal - y_val
        valLoss = ((errorVal**2).mean())
        valLosses.append(valLoss)

        # Plotting (Only updating curves for brevity, can enable scatters if needed)
        plt.figure('SGD Train Loss')
        plt.plot(trainLosses, label=f'LR={lr}')

    plt.legend()
    plt.title("SGD Training Loss Curves")
    plt.xlabel("Epoch")
    plt.ylabel("MSE Loss")
    plt.show(block=False)
    plt.pause(0.1)

plt.show()
```

```python
# MINI-BATCH GRADIENT DESCENT

learning_rates = [0.001, 0.01, 0.05, 0.1, 0.5]
batch_size = 16  # Defining batch size

print("\nSTARTING MINI-BATCH GRADIENT DESCENT")

for lr in learning_rates:
    print(f"\n=== Running Mini-Batch (Size={batch_size}) for Learning Rate = {lr} ===")

    trainLosses = []
    valLosses = []

    w = np.random.randn(1)
    b = np.random.randn(1)

    for i in range(100): # Epochs

        # Shuffle data
        perm = np.random.permutation(len(x_train))
        x_shuffled = x_train[perm]
        y_shuffled = y_train[perm]

        # Iterate via Batches
        for j in range(0, len(x_train), batch_size):
            # Create batch
            x_batch = x_shuffled[j : j + batch_size]
            y_batch = y_shuffled[j : j + batch_size]

            # Forward pass (batch)
            yhat_batch = w * x_batch + b
            error_batch = yhat_batch - y_batch

            # Gradients (average over batch)
            db = 2 * error_batch.mean()
            dw = 2 * (x_batch * error_batch).mean()

            # Update
            b = b - lr * db
            w = w - lr * dw

        # --- Calculate Full Epoch Loss for plotting ---
        yhat_full = w * x_train + b
        loss = ((yhat_full - y_train)**2).mean()
        trainLosses.append(loss)

        # Validation
        yhatVal = w * x_val + b
        valLoss = ((yhatVal - y_val)**2).mean()
        valLosses.append(valLoss)

        plt.figure('MiniBatch Train Loss')
        plt.plot(trainLosses, label=f'LR={lr}')

    plt.legend()
    plt.title(f"Mini-Batch (Size {batch_size}) Training Loss Curves")
    plt.show(block=False)
    plt.pause(0.1)

plt.show()
```
